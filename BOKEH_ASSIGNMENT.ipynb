```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
# Import necessary libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Set output to notebook (can also use output_file('filename.html') for standalone HTML)
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 10, 100)  # x values from 0 to 10
y = np.sin(x)                 # y values as the sine of x

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

# Add a line renderer for the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, line_color="blue")

# Show the plot
show(p)
```

```python
#2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
# Import necessary libraries
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set output to notebook (can also use output_file('filename.html') for standalone HTML)
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(10, 100, n)  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors

# Create a DataFrame to organize the data
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)

# Create a scatter plot
p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add scatter renderer
p.scatter('x', 'y', size='sizes', color='colors', legend_field='colors', source=df, fill_alpha=0.6)

# Show the plot
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
#fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd

# Enable output in a Jupyter notebook (comment this line if not in Jupyter)
output_notebook()

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
df = pd.DataFrame({'Fruits': fruits, 'Counts': counts})

# Create a Bokeh figure
p = figure(x_range=df['Fruits'], height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Create the bar chart
p.vbar(x='Fruits', top='Counts', width=0.9, source=df, color='blue')

# Formatting the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.xaxis.major_label_orientation = "vertical"

# Show the plot
show(p)
```

```python
#4 Create a Bokeh histogram to visualize the distribution of the given data.
#data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30)
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable output in a Jupyter notebook (comment this line if not in Jupyter)
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           tools="")

# Create the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black')

# Show the plot
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
#data_heatmap = np.random.rand(10, 10) x = np.linspace(0, 1, 10) y = np.linspace(0, 1, 10) xx, yy = np.meshgrid(x, y)
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.palettes import Viridis256

# Enable output in a Jupyter notebook (comment this line if not in Jupyter)
output_notebook()

# Generate random data
data_heatmap = np.random.rand(10, 10)

# Create grid for x and y
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for plotting
data = data_heatmap.flatten()
x_coords = xx.flatten()
y_coords = yy.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data={'value': data, 'x': x_coords, 'y': y_coords})

# Create a Bokeh figure
p = figure(title="Heatmap of Random Data",
           x_axis_label='X Axis',
           y_axis_label='Y Axis',
           tools="",
           width=400, height=400)

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data), high=max(data))

# Create a heatmap using quad
p.quad(top='value', bottom=0, left='x', right='x',
        line_color='white', fill_color=mapper, source=source)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
